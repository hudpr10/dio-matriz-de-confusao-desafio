```python
# Acurácia - Classe A
total_acertos_a = 17
total_itens_a = 17 + 18 + 11

acc = total_acertos_a / total_itens_a

print(f"Acurácia classe A = {round(acc*100, 2)}%")
```

```python
# Sensibilidade - Classe A
vp = 17
fn = 18 + 11

sensibilidade = vp / (vp + fn)

print(f"Sensibilidade para classe A = {round(sensibilidade*100, 2)}%")
```

```python
# Especificidade - Classe A
vn = 11 + 3
fp = 15 + 9

especificidade = vn / (fp + vn)
print(f"Especificidade para classe A = {round(especificidade*100, 2)}%")
```

```python
# Precisão - Classe A
vp = 17
fp = 15 + 9

precisao = vp / (vp + fp)
print(f"Precisão para classe A = {round(precisao*100, 2)}%")
```

```python
# F-Score - Classe A
p = 41.46
s = 36.96

f_score = 2 * (p * s) / (p + s)
print(f"F-Score para Classe A = {round(f_score, 2)}%")
```
